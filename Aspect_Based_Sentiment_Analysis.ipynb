# ASPECT BASED SENTIMENT ANALYSIS

## Description- 
This script leverages the PyABSA library for aspect-based sentiment analysis. It uses pre-trained models to extract aspects and sentiments from input sentences. The results are then stored in a pandas DataFrame and saved to an Excel file with conditional formatting to highlight positive and negative sentiments.

```python
#Showing all checkpoints available in PyAbsa
from pyabsa import available_checkpoints

checkpoint_map = available_checkpoints(show_ckpts=True)

print(checkpoint_map)
```

```python
#Choosing the checkpoint most suited to our purpose
from pyabsa import ATEPCCheckpointManager
aspect_extractor = ATEPCCheckpointManager.get_aspect_extractor(checkpoint='english')
```

```python
#Converting our dataset to a usable list
import pandas as pd

df = pd.read_csv('/content/Evaluation-dataset.csv', header=None)

first_column = df[0].tolist()

print(first_column[0:5])
```

```python
#to test our model on the dataset provided
inference_source = first_column
atepc_result = aspect_extractor.extract_aspect(inference_source=inference_source,
                                               pred_sentiment=True)
```

```python
#printing the results
atepc_result[0:10]
```

```python
#extracting the sentence, aspect, sentiment and confidence scores from our results
def extract_aspects(data):
    extracted_data = []
    for entry in data:
        sentence = entry['sentence']
        for aspect, sentiment, confidence in zip(entry['aspect'], entry['sentiment'], entry['confidence']):
            extracted_data.append({
                'sentence': sentence,
                'aspect': aspect,
                'sentiment': sentiment,
                'confidence': confidence
            })
    return extracted_data

extracted_data = extract_aspects(atepc_result)
```

```python
#storing results into a pandas dataframe
import pandas as pd
df = pd.DataFrame(extracted_data)
print(df)
```
